**PART 2 -  Visualization of Data using matplotlib, pyplots Pack
ages[CO2]**
 1. Generate a Data set for Health Care using the following

```python
import matplotlib.pyplot as plt
```

```python
people= ["chodhitha","charvika","shivani","mahesh"]
age =[18,17,19,20]
height =[153,145,140,173]
weight=[42,40,45,57]
```

```python
plt.scatter(weight,height)
plt.title("relation of height and weigth of the patients")
plt.xlabel("weight of patients in kg")
plt.ylabel("height in meters")
plt.show()
```

**Bar plot**

```python
plt.figure(figsize=(13,6))
plt.title("weight in kgs",fontsize=20,fontstyle='oblique')
plt.bar(x=people,height=weight,width=0.3,color='pink',alpha=0.9)
plt.xlabel("people",fontsize=15)
plt.ylabel("weight in kgs",fontsize=15)
plt.xticks(fontsize=16)
plt.yticks(fontsize=18)
plt.show()
```

**HISTOGRAM**

```python
plt.figure(figsize=(6,6))
plt.title("Histogram of patients weight",fontsize=16,fontstyle='oblique')
plt.hist(weight,color='green',edgecolor='k',alpha=0.7,bins=5)
plt.xlabel("weight in kgs",fontsize=15)
plt.xticks(fontsize=16)
plt.yticks(fontsize=18)
plt.show()
```
